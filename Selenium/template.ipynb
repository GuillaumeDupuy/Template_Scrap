Import 

```python
import time
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests 
import os 
```

Avoir le pwd du dossier universel

```python
cwd = os.getcwd()
```

Générer notre page Web

```python
browser = webdriver.Chrome(cwd + '/ChromeDriver/chromedriver')
```

Donner un url à notre page Web

```python
browser.get('https://www.nintendo.fr/Rechercher/Rechercher-299117.html?f=147394-86')
```

Mettre la page en plaine écran

```python
browser.maximize_window()
time.sleep(2)
```

Récupérer les infos qu'on veut

```python
ListModeleDF = []
ListModeleElement = []

for articles in range(1,25):
    
    name = browser.find_element_by_xpath("/html/body/div[4]/div[3]/div/div[1]/div/div[2]/div[2]/div/div/div[2]/ul/li["+str(articles)+"]/div/div/a/div[2]/p[1]/span")
    price = browser.find_element_by_xpath("/html/body/div[4]/div[3]/div/div[1]/div/div[2]/div[2]/div/div/div[2]/ul/li["+str(articles)+"]/div/div/a/div[2]/div[1]/p[2]/span")
    desc = browser.find_element_by_xpath("/html/body/div[4]/div[3]/div/div[1]/div/div[2]/div[2]/div/div/div[2]/ul/li["+str(articles)+"]/div/div/a/div[2]/p[2]")
    data = browser.find_element_by_xpath("/html/body/div[4]/div[3]/div/div[1]/div/div[2]/div[2]/div/div/div[2]/ul/li["+str(articles)+"]/div/div/a/div[2]/div[1]/p[1]")
    plateforme = str(data.text).split(" • ")[0]
    date_release = str(data.text).split(" • ")[1]
    genre = str(data.text).split(" • ")[2]

    
    ListModeleElement.append(str(name.text))
    ListModeleElement.append(str(price.text))
    ListModeleElement.append(str(desc.text))
    ListModeleElement.append(str(plateforme))
    ListModeleElement.append(str(date_release))
    ListModeleElement.append(str(genre))
    ListModeleDF.append(ListModeleElement[:])
    ListModeleElement.clear()
    # print(name.text,price.text,desc.text,plateforme.text,date_release.text,genre.text)

browser.close()
```

Créer un csv de nos infos

```python
df = pd.DataFrame(ListModeleDF)    
df.columns += 1 
df.to_csv("game_nintendo.csv")
```

Lire notre csv et modifier le nom des colonnes

```python
df = pd.read_csv("game_nintendo.csv", encoding='utf-8')
df = df.drop(columns=['Unnamed: 0'])
df = df.rename(columns={'1': 'Name', '2': 'Price', '3': 'Description', '4': 'Plateforme', '5': 'Date Realease', '6': 'Genre'})
```

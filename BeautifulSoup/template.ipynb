```python
from bs4 import BeautifulSoup
import pandas as pd
import requests 
import os 
```

```python
cwd = os.getcwd()
```

```python
url="https://realpython.github.io/fake-jobs/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
results = soup.find(id="ResultsContainer")
```

```python
# Création d'une liste vide pour stocker les données
data = []

# On récupère les informations de chaque annonce

job_elements = results.find_all("div", class_="card-content")

for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    link_url = job_element.find_all("a")[1]["href"]

    # On stocke les données dans la liste data
    data.append([title_element.text.strip(), company_element.text.strip(), location_element.text.strip(), link_url])


# On crée un dataframe à partir de la liste data
df = pd.DataFrame(data, columns=['Titre', 'Company', 'Location', 'Link_url'])

# On sauvegarde le dataframe dans un fichier csv
df.to_csv(cwd + '/jobs.csv', index=False, encoding='utf-8')

df
```

```python
# Création d'une liste vide pour stocker les données
data = []

# On récupère les informations de chaque annonce qui contient le mot "python"
python_jobs = results.find_all("h2", string=lambda text: "python" in text.lower())

python_job_elements = [h2_element.parent.parent.parent for h2_element in python_jobs]

for job_element in python_job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    link_url = job_element.find_all("a")[1]["href"]
    
    # On stocke les données dans la liste data
    data.append([title_element.text.strip(), company_element.text.strip(), location_element.text.strip(), link_url])

# On crée un dataframe à partir de la liste data
df = pd.DataFrame(data, columns=['Titre', 'Company', 'Location', 'Link_url'])

# On sauvegarde le dataframe dans un fichier csv
df.to_csv(cwd + '/jobs_python.csv', index=False, encoding='utf-8')

df
```
